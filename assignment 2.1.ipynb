```python

qua_unique = pd.DataFrame(columns=['Quaters','flavor'])

for i in range(1,5):
    df_i= df[df['quarter']==i]

    fl_i=df_i['flavor'].dropna()
    flavors_i=[]
    for h in fl_i:
        flavors_i.append(h.split(';'))
    for x in range(len(flavors_i)):
        for y in range(len(flavors_i[x])):
            if "||" in flavors_i[x][y]:
                flavors_i[x][y]=flavors_i[x][y].split('||')[-1]
                
    unique_flavors_i=set()
    for o in flavors_i:
        for p in o:
            a = p.split(',')
            for u in range(len(a)):
                a[u]=a[u].title()
                a[u]=a[u].strip()
            
            unique_flavors_i.add(", ".join(a))     
    
    for k in unique_flavors_i:

        qua_unique=qua_unique.append({'Quaters': i , 'flavor' : k}, ignore_index=True)

    
```
